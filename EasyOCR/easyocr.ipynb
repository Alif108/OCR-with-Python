```python
import easyocr
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

IMAGE_PATH = '/content/2.jpg'

reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH)

for i in range(len(result)):
  print(result[i][-2], end=" ")
```
