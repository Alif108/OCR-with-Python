```python
import os
from PIL import Image
import pytesseract
import numpy as np
from pytesseract import Output
import cv2
import matplotlib.pyplot as plt
```

```python
tesseract_path = os.path.join("", 'D:\\Tesseract\\tesseract.exe')
pytesseract.pytesseract.tesseract_cmd = tesseract_path
```

```python
img_path = 'Samples/9.png'

# show image
img = cv2.imread(img_path)
plt.imshow(img)
plt.show()
```

Note : Let's say you calculate the 90th percentile confidence as 80. It means that 90% of the individual confidence scores are 80 or higher

```python
text = pytesseract.image_to_string(img)
result = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)
print(text)
print("Confidence: ", result['conf'])
print("the 90th percentile confidence: ", np.percentile(result['conf'], 90))
```

## Histogram Equalize

```python
img = cv2.imread(img_path)

# Convert to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
enhanced_img = cv2.equalizeHist(gray_img)

plt.imshow(enhanced_img, cmap='gray')
plt.show()
```

```python
text = pytesseract.image_to_string(enhanced_image)
result = pytesseract.image_to_data(enhanced_image, output_type=pytesseract.Output.DICT)
print(text)
print("Confidence: ", result['conf'])
print("the 90th percentile confidence: ", np.percentile(result['conf'], 90))
```

## Check brightness

```python
img = cv2.imread(img_path)

# Convert to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Calculate the mean brightness
mean_brightness = gray_image.mean()
print("Mean brightness:", mean_brightness)

# Threshold to create a binary img
threshold = 150
binary_image = gray_image > threshold

# show the binary img
plt.imshow(binary_image, cmap='gray')
plt.show()
```

```python
text = pytesseract.image_to_string(binary_image)
result = pytesseract.image_to_data(binary_image, output_type=pytesseract.Output.DICT)
print(text)
print("Confidence: ", result['conf'])
print("the 90th percentile confidence: ", np.percentile(result['conf'], 90))
```

## Blur Detection

```python
image = cv2.imread(img_path)

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Calculate the Laplacian variance as a measure of image sharpness
laplacian_var = cv2.Laplacian(gray_image, cv2.CV_64F).var()
print("Laplacian variance:", laplacian_var)

laplacian_var_threshold = 100

if laplacian_var > laplacian_var_threshold:
    print("Image is sharp")
else:
    print("Image is blurry")
```
